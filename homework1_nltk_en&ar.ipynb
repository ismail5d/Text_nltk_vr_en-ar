```python
#import packages
import nltk
import matplotlib
from nltk.stem.isri import ISRIStemmer
from nltk.corpus import treebank
from nltk.corpus import wordnet
from nltk.corpus import stopwords, state_union
from nltk.tokenize import sent_tokenize, word_tokenize,PunktSentenceTokenizer
from nltk.tokenize import PunktSentenceTokenizer #unsupervised machine learning tokenizer(you can train it if you want !)
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
#tokenization by word
example_text1="""ربما كانت أحد أهم التطورات التي قامت بها الرياضيات العربية التي بدأت في هذا الوقت بعمل الخوارزمي و هي بدايات الجبر، و من المهم فهم كيف كانت هذه الفكرة الجديدة مهمة، فقد كانت خطوة ثورية بعيدا عن المفهوم اليوناني للرياضيات التي هي في جوهرها هندسة، الجبر كان نظرية موحدة تتيح الأعداد الكسرية و الأعداد اللا كسرية، و قدم وسيلة للتنمية في هذا الموضوع مستقبلا. و جانب آخر مهم لإدخال أفكار الجبر و هو أنه سمح بتطبيق الرياضيات على نفسها بطريقة لم تحدث من قبل"""
example_text = """Perhaps one of the most significant advances made by Arabic mathematics began at this time with the work of al-Khwarizmi, namely 
the beginnings of algebra. It is important to understand just how significant this new idea was. It was a revolutionary move away from 
the Greek concept of mathematics which was essentially geometry. Algebra was a unifying theory which allowed rational 
numbers, irrational numbers, geometrical magnitudes, etc., to all be treated as "algebraic objects". It gave mathematics a whole new 
development path so much broader in concept to that which had existed before, and provided a vehicle for future development of the 
subject. Another important aspect of the introduction of algebraic ideas was that it allowed mathematics to be applied to itself in a 
way which had not happened before.
"""
#english version
print(word_tokenize(example_text))
print()
#arabic version
print(word_tokenize(example_text1))
```

```python
#print each word individually english
for i in word_tokenize(example_text):
    print(i)
#print each word individually arabic
for i in word_tokenize(example_text1):
    print(i)
```

```python
#tokenization by sentence english
print(sent_tokenize(example_text))
print()
#tokenization by sentence arabic
print(sent_tokenize(example_text1))
```

```python
#print each sentence individually
for i in sent_tokenize(example_text):
    print(i)
```

```python
#stop words in english just to get rid off useless words who's not gonna effect the meaning 
stop_words = set(stopwords.words("english"))
filtered_sentence = []
words = word_tokenize(example_text)
for w in words:
    if w not in stop_words:
        filtered_sentence.append(w)
        
print(filtered_sentence)
print()
#stop words in arabic just to get rid off useless words who's not gonna effect the meaning 
stop_words1 = set(stopwords.words("arabic"))
filtered_sentence = []
words1 = word_tokenize(example_text1)
for w in words1:
    if w not in stop_words1:
        filtered_sentence.append(w)
        
print(filtered_sentence)
```

```python
#Stemming with nltk(you can notice easily how stemming is not that efficient)english
ps = PorterStemmer()
for w in words:
    print(ps.stem(w))
    
print()
#Stemming with nltk(you can notice easily how stemming is not that efficient)arabic
st = ISRIStemmer()
for w in words1:
    print(st.stem(w))
```

```python
#Part of Speech Tagging based on 2005-GWBush text 
train_text = state_union.raw("2005-GWBush.txt")  #training text
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)
#other way of tokenization
tokenized = custom_sent_tokenizer.tokenize(example_text)

def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            print(tagged)
    except Exception as e:
        print(str(e))
        
process_content()
```

```python
#chunking example
def chunking():
    for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            chunkGram = r"""chunk: {<RB.?>*<VB.?>*<NNP><NN>?} """
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            #print(chunked)
            chunked.draw()#using matplotlib to plot the result or just simply by using print(chunked)
            
            
chunking()
```

```python
#Named Entity Recognition english
def named_en():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            namedEn = nltk.ne_chunk(tagged)
            namedEn.draw()
    except Exception as e:
        print(str(e))
        
named_en()
```

```python
#Lemmatization
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
# 1. Init Lemmatizer
lemmatizer = WordNetLemmatizer()

# 2. Lemmatize Single Word with the appropriate POS tag
word = 'feet'
print(lemmatizer.lemmatize(word, get_wordnet_pos(word)))

# 3. Lemmatize a Sentence with the appropriate POS tag

print([lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in nltk.word_tokenize(example_text)])
```
